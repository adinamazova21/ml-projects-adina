```python
#Exercise 2

advertising_spend = [1500, 1700, 1600, 2000, 2100,
                     2500, 2300, 2700, 2800, 3000,
                     3200, 3300, 3400, 3600, 3700,
                     3800, 3900, 4000, 4100, 4200]

sales = [12000, 13000, 14000, 18000, 22000,
         25000, 23000, 30000, 32000, 35000,
         40000, 42000, 45000, 50000, 52000,
         55000, 58000, 60000, 63000, 65000]
```

```python
#Exercise 3

#Part A)

months = list(range(1, 13))
total_profit = [211000, 183300, 224700, 222700,
                209600, 201400, 295500, 361400,
                234000, 266700, 412800, 300200]
```

```python
#creating a line graph, total profit on x-axis, months on y-axis
plt.plot(months, total_profit, color='green', marker='o', linestyle='solid')

plt.title("Total Profit Over 12 Months") #adding a title
plt.xlabel("Month") #labeling the x-axis
plt.ylabel("Total Profit ($)")  #labeling the y-axis

#showing the plot
plt.show()
```
