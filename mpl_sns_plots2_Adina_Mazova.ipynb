```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("classic")
import seaborn as sns
```

# Задача №3:
3.1. Импортируйте и ознакомьтесь с датасетом flights

3.2. Выведите на экран вертикальную столбчатую диаграмму, в которой определена зависимость Года и Количества пассажиров. Данные должны быть разделены по месяцам


```python
#3.1
flights = sns.load_dataset("flights")
flights.head()
```

```python
#3.2
sns.barplot(x = 'year', y = 'passengers', hue = 'month', data = flights).set(
    xlabel = "Год", ylabel = "Количество пассажиров", title = "Вертикальная столбчатая диаграмма Года и Количества пассажиров по месяцам"
)
```

# Задача №4:
4.1. Импортируйте и ознакомьтесь с датасетом diamonds

4.2. Выведите на экран диаграмму рассеяния, в которой определена зависимость Количества карат и Цены за бриллиант

4.3. Выведите на экран диаграммы рассеяния, разделенных по критерию качества бриллиантов (столбец ‘cut’) для зависимости Цены и Количества карат


```python
#4.1
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
#4.2
sns.scatterplot(x = "carat", y = "price", data = diamonds).set(
    xlabel = "Количество карат", ylabel = "Цена за бриллиант", title = "Диаграмма рассеяния Количества карат и Цены за бриллиант"
)
```

```python
#4.3
sns.scatterplot(x = "price", y = "carat", hue = "cut", data = diamonds).set(
    xlabel = "Цена", ylabel = "Количество карат", title = "Диаграмма рассеяния Цены и Количества карат и cut"
)
```
